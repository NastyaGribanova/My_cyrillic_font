```python
import os
import torch
from skimage.util import random_noise
from PIL import Image
import numpy as np

images_path = 'Images'
for dir in os.listdir(images_path):
    if (str(dir) != ".ipynb_checkpoints"):
        aux_dir = os.path.join(images_path,dir)
        
        if os.path.isdir(aux_dir):
            for file in os.listdir(aux_dir):
                filename = os.fsdecode(file)
                if filename.endswith("10.jpg"):
                    im = Image.open(aux_dir+ "/"+filename)
                    im_arr = np.asarray(im)
                    salt_img = random_noise(im_arr, mode='salt', amount=0.05)
                    new_path = "Salt" +  "/" + (filename[:len(filename)-4]) + 'salt10.jpg'
                    salt_img = (255*salt_img).astype(np.uint8)
                    img = Image.fromarray(salt_img)
                    print(new_path)
                    img.save(new_path)
```
