```python
import cv2
import numpy as np
import easyocr
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
my_img_path = '../models_panagram/rgb/Images/Inter/Inter-Black1.jpg'
```

```python
def recognize_text(img_path):   
    reader = easyocr.Reader(['ru'])
    return reader.readtext(img_path)
```

```python
result = recognize_text(my_img_path)
for (bbox, text, prob) in result:
    (top_left, top_right, bottom_right, bottom_left) = bbox
```

```python
example = ''
for (_, t, _) in result:
    example = example + t + ' '
```

```python
example
```

```python
left = top_left[0]
upper = top_left[1]
right = bottom_right[0]
lower = bottom_right[1]

box=(left, upper, right, lower)
```

```python
im = Image.open(my_img_path)
im_crop = im.crop(box)
```

```python
im_crop
```

```python
import skimage
import PIL
def to_gray(img):
    img_array = np.array(img)
    img = skimage.color.rgb2gray(img_array) * 255
    pp_my_img = PIL.Image.fromarray(img).convert('RGB')
    #pp_my_img = image.img_to_array(pp_my_img)
    #pp_my_img = pp_my_img/255    
    return pp_my_img
```

```python
im_crop = to_gray(im_crop)
```

```python
img_grey = cv2.imread('Images/img2.png')

thresh = 200

# threshold the image
img_binary = cv2.threshold(img_grey, thresh, 255, cv2.THRESH_BINARY)[1]

#save image
cv2.imwrite('black-and-white.png',img_binary) 
```

```python
im_crop.save('Images/img2.png')
```

```python
width, height = im_crop.size
```

```python
coordinate = x, y = 0, 0
colors = []
for x in range(width):
    coordinate_cur = x, y = x, y
    current_color = im_crop.getpixel(coordinate_cur)
    if current_color not in colors:
        colors.append(current_color)
        
coordinate = x, y = 0, 0
for y in range(height):
    coordinate_cur = x, y = x, y
    current_color = im_crop.getpixel(coordinate_cur)
    if current_color not in colors:
        colors.append(current_color)
```

```python
colors
```

```python
coordinate = x, y = 0, 0
for y in range(height):
    for x in range(width):        
        coordinate_cur = x, y = x, y
        current_color = im_crop.getpixel(coordinate_cur)
        if current_color not in colors:
            im_crop.putpixel((x, y), (0, 0, 0, 255))
        else:
            im_crop.putpixel((x, y), (255, 255, 255, 255))
```
