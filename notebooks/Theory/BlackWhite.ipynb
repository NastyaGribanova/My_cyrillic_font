```python
import skimage
import PIL
import numpy as np

def to_gray(img):
    img_array = np.array(img)
    img = skimage.color.rgb2gray(img_array) * 255
    pp_my_img = PIL.Image.fromarray(img).convert('RGB')
    #pp_my_img = image.img_to_array(pp_my_img)
    #pp_my_img = pp_my_img/255    
    return pp_my_img
```

```python
import cv2

#read image as grey scale
img_grey = cv2.imread('../models_panagram/rgb/Images/Inter/Inter-Black1.jpg', cv2.IMREAD_GRAYSCALE)

# define a threshold, 128 is the middle of black and white in grey scale
thresh = 128

# threshold the image
img_binary = cv2.threshold(img_grey, thresh, 255, cv2.THRESH_BINARY)[1]

#save image
cv2.imwrite('black-and-white.png',img_binary) 
```
